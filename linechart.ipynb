```python
import seaborn as sns
import numpy as np
import pandas as pd


sns.set(style='white')

```

```python
x = np.array([20, 50, 100, 200, 300])
y = np.array([45, 86, 98, 114, 122])

d = {'x': x, 'y': y}
df = pd.DataFrame(d)
ax = sns.lineplot(x='x', y='y', data=df)
ax.set(xlabel="Number of atoms(k)",ylabel="Time(sec)")
fig = ax.get_figure()
fig.savefig("figs/time_images.pdf")
```
